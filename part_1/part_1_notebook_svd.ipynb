```python
data = []

with open('jester-data-1.csv', 'rb') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for i, row in enumerate(spamreader):
        data.append(row)
        if i == 9999:
            break
```

```python
for i in range(1, len(data)):
    for j in range(len(data[i])):
        if (float(data[i][j]) > 10) or (float(data[i][j]) < -10):
            data[i][j] = 0

for i in range(len(data)):
    if i == 0:
        data[i] = ["users"] + data[i]
    else:
        data[i] = [str(i)] + data[i]

import numpy as np

data = np.asarray(data)


print (data)
```

```python
import surprise
from surprise import SVD
from surprise import Dataset
from surprise import evaluate, print_perf
from surprise.evaluate import GridSearch

import pandas as pd


data = pd.DataFrame(data=data, index=data[0:,0], columns=data[0,0:])

data = pd.melt(data, 
            id_vars='users', 
            value_vars=list(data.columns[1:]),
            var_name='joke', 
            value_name='rating')


reader = surprise.dataset.Reader(rating_scale=(-10, 10))
## The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(data, reader=reader)

data.split(n_folds=3)

algo = SVD()


param_grid = {'n_factors': [20, 100], 'n_epochs': [5, 10], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}

grid_search = GridSearch(SVD, param_grid, measures=['RMSE', 'MAE'])

grid_search.evaluate(data)
```
